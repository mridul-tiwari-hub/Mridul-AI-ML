# Canada Per Capita Income Prediction

This notebook builds a simple linear regression model to predict Canada's per capita income in the year 2020 based on historical data.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
print('Coefficient (m):', reg.coef_[0])
print('Intercept (b):', reg.intercept_)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
```

```python
# Predict Canada's per capita income in year 2020
reg.predict([[2020]])
```
